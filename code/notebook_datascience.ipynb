### Pandas

* Think of pandas like using SQL, but in python
* It allows to work with tabular data in an easy way
* Tabular data - Excel - rows and columns

* Underneath, pandas uses **Numpy**
* Numpy is a high-performance library to work with numerical data
* Usually numpy is used to work with matrices
* But pandas offers nice things to work with strings as well

```python
import pandas as pd
import numpy as np
```

```python
names = [column[0] for column in cursor.description]
```

```python
df = pd.DataFrame(rows, columns=names)
```

```python
df
```

```python
df.sort_values('name')[:10]
```

```python
life_exp_sorted = df.sort_values('lifeexpectancy', ascending=False)[:10]
life_exp_sorted
```

```python
import matplotlib.pyplot as plt
```

```python
life_exp_sorted.plot.bar(x='name', y='lifeexpectancy', rot=0)
```
